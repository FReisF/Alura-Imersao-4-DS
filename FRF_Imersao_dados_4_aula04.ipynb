```python
#cria na tabela inderco uma cola juntado o tupo e o nome do logradouro para se tornar a colua 'RUa'
enderecos_sp["rua"] = enderecos_sp["tipo_logr"] + " " + enderecos_sp["logr_nome"]
#Normaliza os nomes de rua para todas as letras em minusculo - importante para evitar mismatch de uppercase com lowercase depois
enderecos_sp["rua"] = enderecos_sp["rua"].str.lower().str.strip()
enderecos_sp.head()
```

```python
dados_geo["Point"] = ""
for i in dados_geo.index:
    dados_geo["Point"][i] = Point(dados_geo["longitude"][i], dados_geo["latitude"][i])

dados_geo['setor_censo'] = dados_geo["Point"].map(
    lambda x: setor_censo_sp.loc[setor_censo_sp.contains(x), 'CD_GEOCODI'].values
).str[0]
dados_geo
```

```python
def mapSP(param,coord):
    nmdict = {"V005":'Rendimento Nominal',"V007":"Rendimento Responsavel","V009":"Rendimento maiores total","V011":"Rendimento Maiores com Renda"}
    nmscore = nmdict[param.split('_')[-1]]
    dados_vendas_censo.plot.scatter(y='latitude', x = 'longitude',s=4,c = dados_vendas_censo[param], alpha = 1,cmap = 'brg',title = f'{nmscore}',ax =ax[coord[0],coord[1]])
```

# 4 - Realizar a análise exploratória e encontrar variáveis relevantes para solução do problema. 

```python
#Correlacao (Linear - Pearson) entre os parametros e o valor do m2
corr_valorm2 = dados_vendas_censo.corr()[['Valor_m2','Valor_anuncio_float']].dropna().sort_values('Valor_m2')
plt.figure(figsize=(8, 12))
sns.heatmap(corr_valorm2, annot = True)
```

  -V009: Valor do rendimento nominal medio de pessoas com mais de 10 anos (total)

  -V011: Valor do rendimento nominal medio de pessoas com mais de 10 anos (com rendimento somente)

  -V005: Valor do rendimento nominal medio de pessoas responsaveis por domicilios particulares (Total)

  -V007: Valor do rendimento nominal medio de pessoas responsaveis por domicilios particulares (com rendimento somente)
