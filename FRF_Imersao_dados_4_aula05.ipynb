### **1 - Adicionar o valor "real" do m2 do imóvel.**

```python
#Valores medios do metro quadrado em Sao Paulo. Obtidos do site: https://www.proprietariodireto.com.br/preco-m2-sao-paulo
media_bairro_sp = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/media_bairro_sp.csv", sep = ',', encoding = 'cp860')
rep = {"τ":"ç","π":"ã","ß":"á","φ":"í","Θ":"é","≤":"ó",'·':'ú'}
media_bairro_sp.replace(rep,regex = True,inplace = True)
media_bairro_sp.head()
```

```python
dados_vendas_censo_nv = pd.merge(dados_vendas_censo, media_bairro_sp, how = 'left', left_on = 'Bairro', right_on = 'Bairro')
dados_vendas_censo_nv
```

```python
bairros_l = list(media_bairro_sp['Bairro'])
m2_mean = dados_vendas_censo_nv[['Bairro','Valor_m2']].groupby('Bairro').mean().reset_index().query('Bairro in @bairros_l')
m2_mean = pd.merge(m2_mean, media_bairro_sp, how = 'left', left_on = 'Bairro', right_on = 'Bairro')
```

## 2 - Testar outro modelos de regressão.

```python
#Lasso Regression
#https://www.statology.org/lasso-regression-in-python/

from numpy import arange
from sklearn.linear_model import LassoCV
from sklearn.model_selection import RepeatedKFold

#define cross-validation method to evaluate model
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

#define model
model = LassoCV(alphas=arange(0, 1, 0.01), cv=cv, n_jobs=-1)

#fit model
model.fit(X_train, Y_train)

#display lambda that produced the lowest test MSE
print(model.alpha_)
```

```python
Y_predict_test = model.predict(X_test)
Y_predict_train = model.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
print('Y teste', Y_predict_test)
print('Y train', Y_predict_train)


print("mae_test: ",mae_test)
print("r2_test: ",r2_test)
print("mae_train ",mae_train)
print("r2_train: ",r2_train)
```

Regressao Lasso nao foi mais edicionente que a regressao linear. Alpha foi baixo sugerindo um modelo bem ruim.Como sugerido a regressao Ridge foi testada.

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
model = RidgeCV(alphas=arange(0, 1, 0.01), cv=cv, scoring='neg_mean_absolute_error')

model.fit(X_train, Y_train)

#display lambda that produced the lowest test MSE
print(model.alpha_)
```

```python
Y_predict_test = model.predict(X_test)
Y_predict_train = model.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
print('Y teste', Y_predict_test)
print('Y train', Y_predict_train)


print("mae_test: ",mae_test)
print("r2_test: ",r2_test)
print("mae_train ",mae_train)
print("r2_train: ",r2_train)
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x=Y_test, y=Y_predict_test)
```

Regressao Ridge tao pouco performou melhor que a linear ou Lasso

## 4 - Trabalhar um pouco mais nos dados e criar novas variáveis.

```python
#Tentar reduzir a discrepancia nos valores da renda. Aplicar a escala Log.
#Calcular densidade de pessoas por m2
#Calcular renda nominal por numero de pessoas
#X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'Vagas','V001', 'V005','V007', 'V009','V011']]
#Y = dados_vendas_censo["Valor_anuncio"]
dados_vendas_censo['V005 log'] = np.log10(dados_vendas_censo['V005'])
dados_vendas_censo['V007 log'] = np.log10(dados_vendas_censo['V007'])
dados_vendas_censo['V0011 log'] = np.log10(dados_vendas_censo['V011'])
dados_vendas_censo["Densidade"] = dados_vendas_censo["V002"]/dados_vendas_censo["Metragem"] 
dados_vendas_censo["Renda per capta"] = dados_vendas_censo["V005"]/dados_vendas_censo["V002"] 
```

```python
plt.figure(figsize=(18, 8))
mask = np.triu(np.ones_like(dados_vendas_censo.corr(), dtype=bool))
heatmap = sns.heatmap(dados_vendas_censo.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlação', fontdict={'fontsize':18}, pad=16);
```

```python
#Retestando o modelo com as novas variaveis + velhas
X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'Vagas','V005 log','V007 log', 'V009 log', 'V0011 log','Densidade','Renda per capta']]
Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr.fit(X_train, Y_train)
Y_predict_test = lr.predict(X_test)
Y_predict_train = lr.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
```

```python
print('Y teste', Y_predict_test)
print('Y train', Y_predict_train)


print("mae_test: ",mae_test)
print("r2_test: ",r2_test)
print("mae_train ",mae_train)
print("r2_train: ",r2_train)
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x=Y_test, y=Y_predict_test)
```

Pouca diferenca com o modelo original. Proxima tentativa: Manter o valor do anuncio para no max 1 000 000.

```python
#Manter valor max de 1 milhao
X = dados_vendas_censo.query("Valor_anuncio <= 1000000")[['Metragem', 'Quartos', 'Banheiros', 'Vagas','V005 log','V007 log', 'V009 log', 'V0011 log','Densidade','Renda per capta']]
Y = dados_vendas_censo.query("Valor_anuncio <= 1000000")["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr.fit(X_train, Y_train)
Y_predict_test = lr.predict(X_test)
Y_predict_train = lr.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
```
